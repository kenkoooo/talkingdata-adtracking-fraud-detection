```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

%matplotlib inline
print('imported')
```

```python
path = '../../data/'
dtypes = {
        'ip'            : 'uint32',
        'app'           : 'uint16',
        'device'        : 'uint16',
        'os'            : 'uint16',
        'channel'       : 'uint16',
        'is_attributed' : 'uint8',
        'click_id'      : 'uint32'
        }
train_columns = ['ip', 'app', 'device', 'os', 'channel', 'click_time', 'is_attributed']
test_columns  = ['ip', 'app', 'device', 'os', 'channel', 'click_time', 'click_id']
```

```python
%%time
train_df = pd.read_csv(path+"train.csv",
#                        nrows=10000000,
                       dtype=dtypes,
                       usecols=train_columns)
```

```python
count_device = train_df.groupby("device", as_index=False).count().sort_values("ip", ascending=False)[["device", "ip"]]
```

```python
count_device.reset_index(inplace=True)
count_device["count_device"] = count_device.ip
count_device = count_device[["device", "count_device"]]
```

```python
count_device["cum"] = count_device.count_device.cumsum() / np.sum(count_device.count_device)
count_device
```

```python
count_device.loc[count_device.cum <= 0.95]
```
